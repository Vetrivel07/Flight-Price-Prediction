```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_excel('Data_Train.xlsx')
```

```python

df_train.head()
```

```python
df_test = pd.read_excel('Test_set.xlsx')
```

```python
df_test.head()
```

```python
final_df = pd.concat([df_train, df_test])
```

```python
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
final_df['Date_of_Journey'].str.split('/').str[0]
```

```python
final_df.head()
```

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
# created seperate columns for Date, Month, and Year
final_df.head()
```

```python
final_df.info()
```

```python
# drop the previous Date_of_Journey column
final_df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
# splitting the 'Arrival_Time' column into two separate columns: 'Arrival_hour' and 'Arrival_min'
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
# convert the data types
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
```

```python
final_df.head()
```

```python
#drop the Arrival_Time column
final_df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df['Dep_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min'] = final_df['Dep_Time'].str.split(':').str[1]

final_df['Dep_hour'] = final_df['Dep_hour'].astype(int)
final_df['Dep_min'] = final_df['Dep_min'].astype(int)

final_df.drop('Dep_Time', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map(
    {'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, 'nan': 1})
```

```python
final_df.head()
```

```python
final_df.drop('Route', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df.info()
```

```python
final_df['Duration_hour'] = final_df['Duration'].str.split(
    ' ').str[0].str.split('h').str[0]
```

```python
final_df[final_df['Duration_hour'] == '5m']
```

```python
final_df.drop(6474, axis=0, inplace=True)
final_df.drop(2660, axis=0, inplace=True)
```

```python
final_df['Duration_hour'] = final_df['Duration_hour'].astype(int)
```

```python
final_df['Duration_hour'] = final_df['Duration_hour']*60
```

```python
final_df['Duration_min'] = final_df['Duration_hour']
```

```python
final_df.drop('Duration_hour', axis=1, inplace=True)
```

```python
final_df.drop('Duration', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = labelencoder.fit_transform(
    final_df['Additional_Info'])
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
# 1. Histogram of Flight Prices

plt.figure(figsize=(10, 6))
sns.histplot(final_df['Price'], bins=30, kde=True, color='blue')
plt.xlabel('Flight Price ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.show()
```

```python
# 2. Scatter Plot of Flight Prices vs. Departure Date

plt.figure(figsize=(10, 6))
sns.scatterplot(data=final_df, x='Date', y='Price', alpha=0.5)
plt.xticks(rotation=0)
plt.xlabel('Departure Date')
plt.ylabel('Flight Price ($)')
plt.title('Flight Prices Over Time')
plt.show()
```

```python
# 3. Bar Chart of Average Flight Prices by Airline
plt.figure(figsize=(12, 6))
mean_prices_airline = final_df.groupby('Airline')['Price'].mean().sort_values()
sns.barplot(x=mean_prices_airline.index, y=mean_prices_airline.values, palette='viridis')
plt.xticks(rotation=45)
plt.xlabel('Airline')
plt.ylabel('Average Flight Price ($)')
plt.title('Average Flight Prices by Airline')
plt.show()

```

```python
# 4. Bar Chart of Average Flight Prices by Departure City
plt.figure(figsize=(12, 6))
mean_prices_departure = final_df.groupby('Source')['Price'].mean().sort_values()
sns.barplot(x=mean_prices_departure.index, y=mean_prices_departure.values, palette='coolwarm')
plt.xticks(rotation=45)
plt.xlabel('Departure City')
plt.ylabel('Average Flight Price ($)')
plt.title('Average Flight Prices by Departure City')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
mean_prices_arrival = final_df.groupby('Destination')['Price'].mean().sort_values()
sns.barplot(x=mean_prices_arrival.index, y=mean_prices_arrival.values, palette='magma')
plt.xticks(rotation=45)
plt.xlabel('Arrival City')
plt.ylabel('Average Flight Price ($)')
plt.title('Average Flight Prices by Arrival City')
plt.show()
```

```python
# 6. Heatmap of Flight Prices by Departure and Arrival Cities

plt.figure(figsize=(12, 8))
price_matrix = final_df.pivot_table(index='Source', columns='Destination', values='Price', aggfunc='mean')
sns.heatmap(price_matrix, cmap='coolwarm', annot=True, fmt='.0f', linewidths=0.5)
plt.xlabel('Arrival City')
plt.ylabel('Departure City')
plt.title('Flight Price Heatmap (Departure vs. Arrival)')
plt.show()
```

```python
# 10. Pie Chart of Highest Price Airlines

plt.figure(figsize=(10, 6))
costly_airlines = final_df.groupby('Airline')['Price'].mean()
plt.pie(costly_airlines, labels=costly_airlines.index, autopct='%1.1f%%', colors=sns.color_palette('dark'), startangle=140)
plt.title('Average Flight Prices by Airline')
plt.show()
```
